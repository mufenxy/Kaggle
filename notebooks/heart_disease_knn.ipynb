```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
heart_disease = pd.read_csv('../datasets/heart.csv')
heart_disease.dropna()
heart_disease.info()
heart_disease.head(5)
```

```python
print(heart_disease.shape)
```

```python
X = heart_disease.drop('target',axis=1)
y = heart_disease['target']
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.3,random_state=42)
```

## 特征工程

需要对分类变量进行编码（注意多分类需要使用独热编码）

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
category_variables = ['cp','restecg','slope','thal']
numerica_variables = ['age','trestbps','chol','thalach','oldpeak']
binery_variables = ['sex','fbs','exang']
preprocessor = ColumnTransformer(
    transformers=[
    ("num",StandardScaler(),numerica_variables),
    ("cat",OneHotEncoder(drop='first'),category_variables),
    ("binary","passthrough",binery_variables),
])
train_X = preprocessor.fit_transform(train_X)
test_X = preprocessor.fit_transform(test_X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(train_X,train_y)
```

```python
knn.score(test_X,test_y)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
logistic_model.fit(train_X,train_y)
```

```python
logistic_model.score(test_X,test_y)
```
