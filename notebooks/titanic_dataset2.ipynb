# 数据基本信息

## 清除缺失值与缺失值填补

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['age'].fillna(df['age'].median(),inplace=True)
df.drop('deck',axis=1,inplace=True)
df.dropna(axis=0,inplace=True)
```

```python
df.head(5)
```

# 数据特征工程

```python
df.drop(['class','who','adult_male','alive','alone'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df['sex'] = df['sex'].map({'female':0,'male':1})
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
```

```python
df.drop('embark_town',axis=1,inplace=True)
numerica_variables = ['age','fare']
category_variables = ['embarked']
bi_or_order = ['pclass','sex','sibsp','parch']
```

```python
y = df['survived']
X = df.drop('survived',axis=1,inplace=False)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('num',StandardScaler(),numerica_variables),
    ('cat',OneHotEncoder(drop='first'),category_variables),
    ('binary','passthrough',bi_or_order)
])
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.fit_transform(X_test)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train,y_train)
```

```python
print(f"准确率：{100*logistic_model.score(X_test,y_test):.1f}%")
```
